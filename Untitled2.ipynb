```python
# âœ… Step 1: Define Applicant as a dictionary
def get_applicant_data():
    age = int(input("Enter your age: "))
    has_accidents = input("Any past accidents? (yes/no): ").strip().lower() == 'yes'
    location = input("Your location (urban/rural): ").strip().lower()

    return {
        'age': age,
        'has_accidents': has_accidents,
        'location': location
    }

# âœ… Step 2: Risk Assessment Function
def assess_risk(applicant):
    risk = 1.0
    if applicant['age'] < 25:
        risk += 0.5
    if applicant['has_accidents']:
        risk += 1.0
    if applicant['location'] == "urban":
        risk += 0.3
    return risk

# âœ… Step 3: Market Adjustment Factor
def get_market_adjustment():
    return 1.05  # Simulated constant for now

# âœ… Step 4: Premium Calculator
def calculate_premium(base_rate, risk_score, market_factor):
    return base_rate * risk_score * market_factor

# âœ… Step 5: Generate Quote
def generate_quote(applicant, premium):
    print("\n--- Insurance Quote ---")
    print(f"Age: {applicant['age']}")
    print(f"Past Accidents: {'Yes' if applicant['has_accidents'] else 'No'}")
    print(f"Location: {applicant['location'].capitalize()}")
    print(f"Final Premium: â‚¹{premium:.2f}")

# âœ… Step 6: Main Function to Run Everything
def main():
    base_rate = 5000.0
    applicant = get_applicant_data()
    risk_score = assess_risk(applicant)
    market_factor = get_market_adjustment()
    premium = calculate_premium(base_rate, risk_score, market_factor)
    generate_quote(applicant, premium)

# ðŸ” Run the program
main()
```
